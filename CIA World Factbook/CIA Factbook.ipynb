```python
import os
import pandas as pd
import sqlite3
import math

os.chdir(r"C:\\Users\sivat\Desktop\Statistics\Python\DataQuest\Data_Analytics\CIA World Factbook")
conn = sqlite3.connect("factbook.db")
c = conn.cursor()
facts = pd.read_sql_query("SELECT * FROM facts", conn)
facts.head(10)
```

```python
facts.shape
```

```python
facts = facts[facts["area_land"] != 0]
facts.shape
```

```python
def proj_pop(df):
    return df["population"]*math.e**(df["population_growth"]*35/100)
    
facts["projected_population"] = facts.apply(proj_pop, axis =1)
facts.head()
    
```

```python
facts_desc_pop = facts.sort_values("projected_population", ascending = False)
facts_desc_pop.head()
```

```python
land_query = "select sum(area_land) from facts;"
water_query = "select sum(area_water) from facts;"

area_land = c.execute(land_query).fetchall()
area_water = c.execute(water_query).fetchall()
```
